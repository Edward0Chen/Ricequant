```python
import tushare as ts
import collections

data1 = ts.get_today_all().sort_values(by='nmc', ascending=False)
print(data1[0: 1000][['code', 'name', 'nmc']])
```

```python
data2 = data1[0: 1000]['code']
data2.index = range(0, 1000)
print(data2)
```

```python
stocks = []
for i in range(0, 1000):
    stocks.append(data2[i])
print(stocks)
```

```python
data3 = ts.get_industry_classified()
data4 = []
for i in range(0, 1000):
    for j in range(0, len(data3)):
        if stocks[i] == data3['code'][j]:
            data4.append(data3.loc[j]['c_name'])
            print(data3.loc[j]['c_name'])
print(len(data4))
data4
```

```python
data5 = collections.Counter(data4)
data5.most_common()
```
